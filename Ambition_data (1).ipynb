```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav',headers = headers).text
```

```python
#webpage
```

## This line is used to convert raw HTML into a readable structure so we can extract data easily.

```python
soup= BeautifulSoup(webpage , 'lxml')
```

```python
#print(soup.prettify())
```

```python
soup.find_all('h2')[0].text.strip()
```

```python
for i in soup.find_all('h2',class_ ="companyCardWrapper__companyName"):
    print(i.text.strip())
len(soup.find_all('h2', class_ ="companyCardWrapper__companyName"))
```

```python
company = soup.find_all('div',class_="companyCardWrapper")
```

```python
len(company)
```

```python
name=[]
rating = []
reviews=[]
salaries = []
interviews = []
jobs = []
info = []

for i in company:
    name.append(i.find('h2', class_ ="companyCardWrapper__companyName").text.strip())
    rating_tag = i.find('div', class_="rating_text rating_text--md")
    if rating_tag:
        rating.append(rating_tag.text.strip())
    else:
        rating.append("No rating")
    reviews.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[0].text.strip())
    salaries.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[1].text.strip())
    interviews.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[2].text.strip())
    jobs.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[3].text.strip())
    info.append(i.find('span', class_="companyCardWrapper__interLinking").text.strip())    
```

```python
df=pd.DataFrame({'Company':name,
   'Rating':rating,
   'Reviews':reviews,
   'Interviews':interviews,
   'Company_Info':info,
   'Jobs':jobs,
   'Salaries':salaries,
   })
```
